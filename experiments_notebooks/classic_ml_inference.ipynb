# Classic ML Inference

This notebook demonstrates how to run inference for one `.data` file using:
- saved CatBoost weights (`catboost_model.cbm`)
- saved preprocessing artifacts (`preprocessing.joblib`)
- CLI script `tools/classic_ml_inference.py`.

````python
from pathlib import Path
import subprocess
import json

project_root = Path.cwd()
if not (project_root / 'tools').exists() and (project_root.parent / 'tools').exists():
    project_root = project_root.parent

script_path = project_root / 'tools' / 'classic_ml_inference.py'
model_path = project_root / 'artifacts' / 'classic_ml' / 'catboost_model.cbm'
preprocessing_path = project_root / 'artifacts' / 'classic_ml' / 'preprocessing.joblib'

data_file = project_root / 'wifi_data_set_fixed' / 'id_person_03' / 'label_02' / 'test_18' / 'test18__dev1_64_E8_33_57_AA_F4.data'

print(f'script: {script_path}')
print(f'model: {model_path}')
print(f'preprocessing: {preprocessing_path}')
print(f'data_file: {data_file}')
````

````python
cmd = [
    'python', str(script_path),
    '--model-path', str(model_path),
    '--preprocessing-path', str(preprocessing_path),
    '--data-file', str(data_file),
    '--pretty',
]

completed = subprocess.run(cmd, capture_output=True, text=True, check=True)
print(completed.stdout)

result = json.loads(completed.stdout)
result
````

## CLI Usage

```bash
python tools/classic_ml_inference.py \
```
